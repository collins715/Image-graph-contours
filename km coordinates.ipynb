```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('km.jpg')

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply a binary threshold to get a binary image
_, binary = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV)

# Find contours
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Separate contours based on color (assuming different colors for each curve)
# Convert image to HSV color space
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define color ranges for each curve (example values, adjust as needed)
# These color ranges were extracted using Adobe Photoshop
lower_red = np.array([0, 20, 0])
upper_red = np.array([10, 255, 255])

# Create masks for each color
mask_red = cv2.inRange(hsv, lower_red, upper_red)

# Find contours for each mask
contours_red, _ = cv2.findContours(mask_red, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Function to map pixel coordinates to graph coordinates
def map_to_graph_coords(x, y, x_axis_start, x_axis_end, y_axis_start, y_axis_end, x_min, x_max, y_min, y_max):
    graph_x = x_min + (x - x_axis_start) * (x_max - x_min) / (x_axis_end - x_axis_start)
    graph_y = y_min + (y_axis_start - y) * (y_max - y_min) / (y_axis_start - y_axis_end)
    return graph_x, graph_y

# Define the pixel positions of the axes
# These positions were extracted using Adobe Photoshop
x_axis_start = 65  
x_axis_end = 465   
y_axis_start = 316
y_axis_end = 58    

# Define the scale of the graph
x_min, x_max = 0, 700
y_min, y_max = 0, 12

# Extract and map coordinates for the red curve
print("Red Curve Coordinates:")
for contour in contours_red:
    for point in contour:
        x, y = point[0]
        graph_x, graph_y = map_to_graph_coords(x, y, x_axis_start, x_axis_end, y_axis_start, y_axis_end, x_min, x_max, y_min, y_max)
        print(f"Graph Coordinate: ({graph_x:.2f}, {graph_y:.2f})")
```
